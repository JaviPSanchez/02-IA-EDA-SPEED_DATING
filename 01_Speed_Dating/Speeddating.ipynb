### Speed Dating Part 1 ###
Start exploring the dataset with pandas to extract some statistics

```python
sd=pd.read_csv('Speed Dating Data.csv', encoding="ISO-8859-1")
```

```python
print("First rows : ")
sd.head()
display(sd.head())
print()
print("statistics : ")
sd.describe(include="all")
display(sd.describe(include="all"))
print()
print("Number of rows : {}".format(sd.shape[0]))
print()
print("Percentage of missing values: ")
display(100*sd.isnull().sum()/sd.shape[0])
print()
print("Details of dataset with info: ")
display(sd.info())
print()

```

```python
# drop columns with isnull>60 %

sd2 = sd.loc[:, 100*sd.isnull().sum()/sd.shape[0]<50]
sd2
```

```python
#control of isnull>60 %

print("Percentage of missing values: ")
display(100*sd2.isnull().sum()/sd2.shape[0])
print()
```

```python
sd2.info(verbose =True)
```

```python
#Selection of columns fr coorelation
columns_ok = ["income","satis_2","exphappy","gender","match","age_o","match_es","int_corr","samerace","race_o","race","dec_o","dec","goal","attr_o","from","field","imprace","imprelig"]
dfsel = sd2.loc[:,columns_ok]
dfsel.info()
```

```python
#renaming of columns
```

```python
#% of missing values and move to float type for "income" column
print("Percentage of missing values: ")
display(100*dfsel.isnull().sum()/dfsel.shape[0])
print("-----------------------")
print()
print("Statistics of selection : ")
dfsel.describe()
# Modification of Income column
print()
print(" Modification of Income column : ")
dfsel['income'] = dfsel['income'].str.replace(',','').astype('float')
print(" Statistics : ")
dfsel.describe(include = 'all')

```

```python
#adding of mean for Nan area
dfsel["income"] = dfsel["income"].fillna(dfsel["income"].mean())
dfsel.describe()
```
